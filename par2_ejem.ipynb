imagina que quieres diseñar un jardin rectangular. la lon del jardin es 3m mas que el ancho. deseas qe el area total del jardin sea 70m cuadrados. diseñe una ec adecuada para el problema y resuelva diseñando metodos de computacion

u=x, 
l=x+3

x*(x+3)=70

x**2+3*x=70

x**2+3*x-70=0

queremos conocer x

1.graficar para conocer intervalos
2.sacar raiz con codigo

```python
from math import *
import matplotlib.pyplot as plt
import numpy as np

t = np.arange(-15.0, 20.0)
s = t**2+3*t-70  
line, = plt.plot(t, s, lw=2)

plt.grid(True)
plt.ylim(-80, 10)

plt.title("$y=x^2+3x-70$")
plt.ylabel('$y = f(x)$')
plt.xlabel('x')
```

```python

def y(x):                    # define la funcion y(x)
  y = x**2+3*x-70
  return y

x1 = float(input('Captura el valor de x1: ')) # peticion de valor x1
x2 = float(input('Captura el valor de x2: ')) # peticion de valor x2
y1 = y(x1)                                    # evalua la funcion y(x1)
y2 = y(x2)                                    # evalua la funcion y(x1)

if y1*y2 > 0:                                 # prueba si los signos son iguales
  print('No hay raices en el intervalo')
  exit

for i in range(100):
  xh = (x1+x2)/2
  yh = y(xh)                                  # evalua la funcion y(xh)
  y1 = y(x1)                                  # evalua la funcion y(x1)
  if abs(y1) < 1.0e-6:
    break
  elif y1*yh < 0:
    x2 = xh
  else:
    x1 = xh
      
print('La raiz es: %.5f' % x1)
print('Numero de bisecciones: %d' % (i+1))
```
